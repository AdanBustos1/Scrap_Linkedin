```python
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument("--disable-extensions")
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--local_state=USA")
driver = webdriver.Chrome(ChromeDriverManager().install(),chrome_options=chrome_options)
```

```python
driver.get("https://linkedin.github.io/career-explorer/")

#Para tener los datos de forma separada por si es necesario
jobs = [] 
all_skills = []
job_skill = dict()
upgrades = dict()
upgrade_skill = dict()
df = pd.DataFrame()



#En total hay 1729 trabajos (lo vi dentro del codigo, no lo conte xD), buscamos uno a uno.
for i in range(1729):
    

    #Vaciar el input de busqueda
    driver.find_element(By.ID, 'explore').find_element(By.ID, 'jobs--ac').send_keys(Keys.CONTROL + 'a', Keys.BACKSPACE)
   
   
    builds = []
    matche = []
    skills = []

    #Buscamos y clickamos a cada resultado de la lista de jobs
    driver.find_element(By.ID, 'explore').find_element(By.ID, 'jobs--ac').send_keys('a') 
    driver.find_element(By.ID, 'jobs--ac__option--'+str(i)).click()


    #Sacamos los job titles
    job = driver.find_element(By.CLASS_NAME, 'title')
    job1 = job.text
    job2 = job1[11:]
    jobs.append(job2)

    #Sacamos las skills
    sk = driver.find_elements_by_xpath("//div[@class ='search__autocomplete']/div[2]/div[2]/ul[1]/li")
    for i in range(len(sk)):
        skills.append(sk[i].text)
        all_skills.append(sk[i].text)

    
    #Sacamos las skills to build
    build = driver.find_elements_by_xpath("//div[@class ='container__jobs']/div[1]/div[2]/div[2]/ul[1]/li")
    for i in range(len(build)):
        builds.append(build[i].text)


    #Sacamos los titulos de puestos superiores
    matches = driver.find_elements_by_xpath("//div[@class ='job__header']/h4[@class ='name']")
    for i in range(len(matches)):
        matche.append(matches[i].text)

    #Lo metemos en diccionarios
    job_skill[job.text] = skills
    upgrades["Upgrade " + job2] = matche
    upgrade_skill["Build "+job2] = builds

    #Creamos un Dataframe
    nueva_fila = pd.Series(data={'Job_title':job2, 'Skills':skills, 'Upgrade_job':matche, 'Skills to upgrade':builds})
    df = df.append(nueva_fila, ignore_index=True)

    
    #Vaciar el input de busqueda
    driver.find_element(By.ID, 'explore').find_element(By.ID, 'jobs--ac').send_keys(Keys.CONTROL + 'a', Keys.BACKSPACE)
    
    
```
